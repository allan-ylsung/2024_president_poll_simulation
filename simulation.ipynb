```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_samples = 1068
n_sims = int(1e7)
p_val = [0.45,0.35,0.2]
data = np.random.multinomial(n_samples,p_val,2*n_sims)
data1, data2 = data[:n_sims]/n_samples, data[n_sims:]/n_samples
```

## 定義
$$(p_0,p_1,p_2)=(柯侯,賴蕭,其他)$$
$$(q_0,q_1,q_2)=(侯柯,賴蕭,其他)$$
$$Prob=假設為真實比率$$

## 95\% CL 柯侯-侯柯 誤差範圍

```python
dif = (data1-data2)[:,0]
ymax = 5.5e5
plt.hist(dif*100,bins=100);
plt.grid()
plt.minorticks_on()
dif_95cl = np.percentile(np.abs(dif),95)*100
plt.plot([-dif_95cl]*2,[0,ymax],'k--',label=f'95%CL={dif_95cl:.2f}%')
plt.plot([dif_95cl]*2,[0,ymax],'k--')
plt.title(f'Samples={n_samples} Prob={p_val}')
plt.xlabel(r'$p_0-q_0$')
plt.legend()
plt.ylim(0,ymax)
```

## 95\% CL (柯侯-賴蕭)-(侯柯-賴蕭) 誤差範圍

```python
dif_dif = (data1[:,0]-data1[:,1])-(data2[:,0]-data2[:,1])
ymax = 5.5e5
plt.hist(dif_dif*100,bins=100);
plt.grid()
plt.minorticks_on()
dif_dif_95cl = np.percentile(np.abs(dif_dif),95)*100
plt.plot([-dif_dif_95cl]*2,[0,ymax],'k--',label=f'95%CL={dif_dif_95cl:.2f}%')
plt.plot([dif_dif_95cl]*2,[0,ymax],'k--')
plt.title(f'Samples={n_samples} Prob={p_val}')
plt.xlabel(r'$(p_0-p_1)-(q_0-q_1)$')
plt.legend()
plt.ylim(0,ymax)
```
